```python
irregular = ['N/A', 'na', 'none', 'None', '?', '-', '--', 'unknown', 'NULL']
for col in df.columns:
    found = df[col].isin(irregular).sum()
    if found > 0:
        print(f"{col}: {found} irregular values found")
    # else:
    #     print(f"No irregular values found")
        
```

```python
print(df.isnull().sum().to_frame('nulls').assign(
    dtype=df.dtypes,
    unique=df.nunique(),
    null_pct=(df.isnull().sum() / len(df) * 100).round(2)
))
```
